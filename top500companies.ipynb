```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns

```

```python
df=pd.read_csv("Top 500 Companies - India.csv",nrows=50)
df.head()
```

```python
print(df.columns)
```

```python
del df['Unnamed: 4']
```

```python
df.head()
```

```python
sns.set(style="ticks", color_codes=True)
```

```python
df.describe()
```

```python
plt.figure(figsize=(10, 8))
sns.barplot(x='Name', y="Mar Cap - Crore", data=df, estimator=np.median)
plt.xticks(rotation =90)
plt.show()
```

```python
plt.figure(figsize=(10, 8))
df.sort_values('Sales Qtr - Crore',ascending=True)
sns.barplot(x='Name', y="Sales Qtr - Crore", data=df, estimator=np.median,)
plt.xticks(rotation =90)
plt.show()
```

```python
df1=pd.read_csv("Top 500 Companies - India.csv")
df1.reset_index()
```

```python
sales=max(df1['Sales Qtr - Crore'])
result=df1.loc[df1['Sales Qtr - Crore']==110666.93,'Name']
result.head()
```

```python
df2=df1.nlargest(50,['Sales Qtr - Crore','Mar Cap - Crore'])
df2.head()
```

```python
plt.figure(figsize=(10, 8))
df2.sort_values('Sales Qtr - Crore',ascending=True)
sns.barplot(x='Name', y='Mar Cap - Crore', data=df2, estimator=np.mean)
plt.xticks(rotation =90)
plt.show()
```

```python
df2['NEWSALES']=df2['Sales Qtr - Crore']/df2['Mar Cap - Crore']
plt.figure(figsize=(10, 8))
df2.sort_values('Sales Qtr - Crore',ascending=True)
sns.barplot(x='Name', y='NEWSALES', data=df2, estimator=np.mean)
plt.xticks(rotation =90)
plt.show()
```
