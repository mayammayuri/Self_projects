```python
account=pd.read_csv("accounts.csv")
clicks= pd.read_csv("clicks.csv")
products=pd.read_csv("products.csv")
sales_pipeline=pd.read_csv("sales_pipeline.csv")
sales_teams=pd.read_csv("sales_teams.csv")
```

```python
sales_pipeline=sales_pipeline.dropna(axis=0,how="any")
sales_pipeline.head
```

Display 'Manager' and 'Grand Total Sales', for sales done by the sales agents reporting these managers

```python
sales_pipeline2=pd.concat([sales_pipeline["sales_agent"],sales_pipeline["close_value"]],axis=1)
output3=pd.merge(sales_teams,sales_pipeline2,on="sales_agent",how="inner")
output3.head
#solution code here
```

```python
fig = px.bar(output3, y='close_value', x='manager', text='close_value')
#fig.update_traces(texttemplate='%{text:.2s}', textposition='outside')
fig.show()
#x=output3["manager"]
#y=output3["close_value"]
#plt.bar(x, y,color ='maroon')
```


 Display 'Sales Agents' and 'Sales' for those sales where product sold at profit

```python
output2=pd.merge(products,sales_pipeline,on="product",how="inner")
output4=output2.query("close_value > sales_price") #for profit
output4['sales'] = output4['close_value'] - output4['sales_price']
output4=output4.query("sales > 0") #for sales
fig2 = px.bar(output4, y='sales', x='sales_agent', text='sales')
fig2.show()
#solution code here
```

 Display the 'Opportunity ID' and 'Days Taken to Close', for opportunities those got closed within a month

```python
sales_pipeline['created_on2'] = pd.to_datetime(sales_pipeline['created_on'], format='%Y-%m-%d')
sales_pipeline['close_date2'] = pd.to_datetime(sales_pipeline['close_date'], format='%Y-%m-%d')
sales_pipeline.drop(['created_on', 'close_date'], axis = 1)
sales_pipeline['Days_to_close'] = ( sales_pipeline['close_date2']- sales_pipeline['created_on2']).dt.days
output5=sales_pipeline.query("Days_to_close < 30") #for profit
sales_pipeline3=pd.concat([output5["opportunity_id"],output5["Days_to_close"]],axis=1)
sales_pipeline3
#fig3 = px.bar(output5, y='Days_to_close', x='opportunity_id', text='Days_to_close')
#fig3.show()
#solution code here
```

Display product(s) got maximum leads (by count) generated from paid source

```python
q41=clicks.loc[(clicks['source'] == 'Paid')] 
q42=sales_pipeline.groupby('created_on2').size().reset_index().rename(columns={0: 'lead_count'})
q422=pd.concat([sales_pipeline['product'],sales_pipeline['created_on2']],axis=1)
q43=pd.merge(q42,q422,on="created_on2",how="inner")
col_max = q43.lead_count.max()
col_max
#q43
#result4 = q43.apply(lambda x: col_max - x['measure'], axis=1)

#solution code here
```

Display in ascending order of revenue, 'Account' and 'Revenue' for telecom accounts 

```python
q61=clicks.query("industry == 'Telecom'")
q62=sales_pipeline.drop(['opportunity_id', 'deal_stage','product'], axis = 1)
q63=pd.merge(q62,q61,on="created_on",how="inner")
q64=pd.merge(q63,account,on="account",how="inner")
q65=pd.concat([q64["revenue"],q64["account"],q64["industry"]],axis=1)
result6=q65.sort_values(by=['revenue'], ascending=[True])
result6=result6.drop_duplicates()

fig6 = px.bar(result6, y='revenue', x='account', text='revenue')
fig6.show()
```

Display 'Month of Year' vs 'Sales', for GTXBasic. NOTE: "Month of Year" means month year (eg. Jan) and "Month" means month (eg. Jan 2020, Jan 2021 etc)

```python
q81=sales_pipeline.query("product == 'GTXBasic'")
output8=sales_pipeline
output8['month_of_year'] = sales_pipeline['close_date2'].dt.strftime('%Y-%m')
output8=output8.groupby('month_of_year').size().reset_index().rename(columns={0: 'count_signups'})
sales = sales_pipeline.loc[(sales_pipeline['product'] == 'GTXBasic')] 
sales=sales.drop(['account', 'opportunity_id','sales_agent','deal_stage','close_date','Days_to_close','created_on'], axis = 1)
#sales
q82=pd.merge(output8,sales,on="month_of_year",how="inner")
#q82
fig8 = px.bar(q82, y='close_value', x='month_of_year', text='close_value')
fig8.show()
#solution code here
```

Display 'Sales Agents', 'Product', and 'Sales', for those sales agents who closed more than one deal on same day

```python
q101=sales_pipeline.groupby('close_date2').size().reset_index().rename(columns={0: 'Days_to_close'})
q101=q101[q101.apply(lambda x: x['Days_to_close']>1, axis=1)]
q102=pd.concat([sales_pipeline["close_value"],sales_pipeline["sales_agent"],sales_pipeline["product"],sales_pipeline['close_date2']],axis=1)
result10=pd.merge(q102,q101,on="close_date2",how="inner")
result10=result10.drop(['Days_to_close', 'close_date2'], axis = 1)
result10
```
